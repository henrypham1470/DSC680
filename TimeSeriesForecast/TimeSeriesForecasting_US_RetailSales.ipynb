# 1.Plot the data with proper labeling and make some observations on the graph.

```python
# Import the pandas library
import pandas as pd

# Read a csv file named 'us_retail_sales.csv' to a pandas dataframe
df = pd.read_csv('us_retail_sales.csv')

# Display the result
df
```

```python
# Use the melt function to convert columns to rows, similarly to pivot function in SQL
melted_df = pd.melt(df, id_vars=['YEAR'], var_name='Month', value_name='Retail_Sales')

# Display the melted_df
melted_df
```

```python
# Import few libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a line plot for monthly retail sales
# Set a size of 12 x 6 inches
plt.figure(figsize=(12, 6))
# Use Seaborn lineplot function to specify the data source (melted_df), 
# the x-axis data ('YEAR'), the y-axis data ('Retail_Sales'), 
# and additional parameters such as hue, palette, and legend. 
# The hue parameter is used to differentiate lines by the 'Month' column, 
# the palette parameter sets the color palette, and legend='full' shows the legend with all categories.
sns.lineplot(data=melted_df, x='YEAR', y='Retail_Sales', hue='Month', palette='husl', legend='full')
# Set the title for the plot.
plt.title('Monthly Retail Sales Over the Years')
# Set the label for the x-axis
plt.xlabel('Year')
# Set the lable for the y-axis
plt.ylabel('Retail Sales')
# Add grid lines to the plot for better read
plt.grid(True)
# Fit the figure
plt.tight_layout()

# Make observations (add text annotations)
plt.text(2010, 420000, 'Growth', fontsize=10, color='black')
plt.text(2008, 300000, 'Short decline', fontsize=10, color='blue')
plt.text(2002, 350000, 'Increase', fontsize=10, color='red')
plt.text(1992, 150000, 'Starting point', fontsize=10, color='brown')

# Show the plot
# Adjust legend position
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
# Display the plot
plt.show()
```

```python
# Here are some observations on the plot with the updated annotations:

# Starting point: The plot begins in 1992, that serves as the starting point for tracking monthly sales trends.
# Increase: Around 2002, there is an increase in monthly sales, denoted as "Increase".
# Short decline: Around 2008, there is a short decline period due to global financial crisis, leading to economic challenges
# Growth: Since 2010, there is a noticeable growth trend in monthly sales.
```

# 2.Split this data into a training and test set. Use the last year of data (July 2020 â€“ June 2021) of data as your test set and the rest as your training set

```python
# Import the pandas library
import pandas as pd

# Combine 'YEAR' and 'Month' columns to create a 'Date' column
melted_df['Date'] = melted_df['YEAR'].astype(str) + '-' + melted_df['Month']

# Convert the 'Date' column to datetime
melted_df['Date'] = pd.to_datetime(melted_df['Date'], format='%Y-%b')

# Set 'Date' as the index
melted_df.set_index('Date', inplace=True)

# Define the test data (last year: July 2020 - June 2021)
test_data = melted_df['2020-07-01':'2021-06-01']

# Define the training data (exclude the last year)
training_data = melted_df[melted_df.index < '2020-07-01']

# Display the shapes of the training and test sets
print("Training data shape:", training_data.shape)
print("Test data shape:", test_data.shape)
```

# 3.Use the training set to build a predictive model for the monthly retail sales.

```python
# Import few libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet

# Combine 'YEAR' and 'Month' columns to create a 'YrMonth' column
# which represents the year and month in a format like 'YYYY-MMM'.
melted_df['YrMonth'] = melted_df['YEAR'].astype(str) + '-' + melted_df['Month']

# Convert the 'YrMonth' column to datetime
# specifying the format '%Y-%b' to match the 'YYYY-MMM' format. This allows to work with dates as datetime objects.
melted_df['YrMonth'] = pd.to_datetime(melted_df['YrMonth'], format='%Y-%b')

# Define the training data (exclude the last year)
# This effectively creates a training dataset that excludes the last year of data.
training_data = melted_df[melted_df.index < '2020-07-01']

# Display the result
training_data
```

```python
# Select two columns: 'YrMonth' and 'Retail_Sales' in training dataset
# pass them to new DataFrame named df_training.
df_training = training_data[['YrMonth','Retail_Sales']]

# Display the result
df_training
```

```python
# Rename YrMonth to ds, datestamp, and Retail_Sales to y.
df_training=df_training.rename(columns={'YrMonth':'ds','Retail_Sales':'y'})

# Display the result
df_training
```

```python
# Create an instance of the Prophet class and assigns it to the variable predict_trainingSales. 
# Prophet is a forecasting tool developed by Facebook for time series forecasting. 
predict_trainingSales=Prophet()

# Fit the Prophet model on the training data. 
# df_training is the DataFrame that contains the historical time series data. 
predict_trainingSales.fit(df_training)
```

```python
# predict_trainingSales: This is the Prophet model instance.
# make_future_dataframe: This method is used to create a DataFrame with dates for the historical training data.
# freq='M': This specifies the frequency of the dates to be on a monthly basis.
pred=predict_trainingSales.make_future_dataframe(periods=-12, freq='M')

# predict: This method is used to make predictions on the DataFrame.
# forecast_train: is a DataFrame that contains the predicted values 
forecast_train = predict_trainingSales.predict(pred)

# [['ds', 'trend']]: This part of the code selects and displays two columns from the forecast_train DataFrame.
# 'ds' which represents the date or time point, and 'trend' which represents the trend component of the forecast.
forecast_train[['ds','trend']]
```

```python
# Import the matplotlib library
import matplotlib.pyplot as plt

# Create the plot
predict_trainingSales.plot(forecast_train, xlabel='Date', ylabel='Retail Sales')

# Set the title for the plot
plt.title('Monthly Retail Sales Forecast (Training data)')

# Adjust legend position
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

# 4.Use the model to predict the monthly retail sales on the last year of data.

```python
# Import few libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet


# Combine 'YEAR' and 'Month' columns to create a 'YrMonth' column
melted_df['YrMonth'] = melted_df['YEAR'].astype(str) + '-' + melted_df['Month']

# Convert the 'YrMonth' column to datetime
melted_df['YrMonth'] = pd.to_datetime(melted_df['YrMonth'], format='%Y-%b')

# Define the training data (exclude the last year)
testing_data = melted_df['2020-07-01':'2021-06-01']

# Display the result
testing_data
```

```python
# Select two columns: 'YrMonth' and 'Retail_Sales' in testing dataset
# pass them to new DataFrame named df_testing.
df_testing = testing_data[['YrMonth','Retail_Sales']]

# Display the result
df_testing
```

```python
# Rename YrMonth to ds, datestamp, and Retail_Sales to y.
df_testing=df_testing.rename(columns={'YrMonth':'ds','Retail_Sales':'y'})

# Display the result
df_testing
```

```python
# Create an instance of the Prophet class and assigns it to the variable predict_testingSales. 
predict_testingSales=Prophet()

# Fit the Prophet model on the testing data. 
# df_testing is the DataFrame that contains the historical time series data.  
predict_testingSales.fit(df_testing)
```

```python
# make_future_dataframe: This method is used to create a DataFrame with dates for the historical testing data.
# freq='M': This specifies the frequency of the dates to be on a monthly basis.
pred_test=predict_testingSales.make_future_dataframe(periods=0, freq='M')

# predict: This method is used to make predictions on the DataFrame.
# forecast_test: is a DataFrame that contains the predicted values 
forecast_test = predict_testingSales.predict(pred_test)

# [['ds', 'trend']]: This part of the code selects and displays two columns from the forecast_test DataFrame.
# 'ds' which represents the date or time point, and 'trend' which represents the trend component of the forecast.
forecast_test[['ds','trend']]
```

```python
# Import the matplotlib library
import matplotlib.pyplot as plt

# Create the plot
predict_testingSales.plot(forecast_test, xlabel='Date', ylabel='Retail Sales')

# Set the title for the plot
plt.title('Monthly Retail Sales Forecast (Testing data)')

# Adjust legend position
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```

# 5. Report the RMSE of the model predictions on the test set

```python
# Import few libraries
from sklearn.metrics import mean_squared_error
import numpy as np

# Extract the actual values from the test set
actual_values = test_data['Retail_Sales'].values

# Extract the predicted values from the forecast DataFrame
# It specifically selects the 'trend' component of the forecast for dates on or after July 1, 2020, 
# and stores these predicted values in the predicted_values variable.
predicted_values = forecast_test.loc[forecast_test['ds'] >= '2020-07-01', 'trend'].values

# Calculate the squared errors
# It subtracts the predicted values from the actual values and squares the result. 
# This step is necessary to compute the MSE.
squared_errors = (actual_values - predicted_values) ** 2

# Calculate the mean squared error (MSE)
# The MSE quantifies the average squared difference between the actual and predicted values.
mse = np.mean(squared_errors)

# Calculate the root mean squared error (RMSE)
# The RMSE is a measure of the standard deviation of the prediction errors and 
# provides a more interpretable measure of the prediction accuracy.
rmse = np.sqrt(mse)

# Print the RMSE
print(f"RMSE: {rmse:.2f}")
```

```python
# The RMSE quantifies how well the model's predictions match the actual data, with lower values indicating better prediction
# accuracy. In this context, it measures how well the Prophet model performs in predicting monthly retail sales on the test data.
```
